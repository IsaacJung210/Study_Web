```python
import folium
import pandas as pd
import requests
import json
import numpy as np
```

```python
#####1번
filename='202202_202202_주민등록인구및세대현황_월간_(1).csv'
df = pd.read_csv(filename,encoding='cp949')
df.head(2)
```

```python
df['행정구역'].head()
```

```python
#####2번
df.rename(columns={'2022년02월_총인구수':'총인구수',
                   '2022년02월_세대수':'세대수',
                  '2022년02월_세대당 인구':'세대당인구',
                  '2022년02월_남자 인구수':'남자',
                  '2022년02월_여자 인구수':'여자',
                  '2022년02월_남여 비율':'남녀비율'},inplace=True)
```

```python
#####3번
df.drop(columns='남녀비율',inplace=True)
```

```python
######4번
df['행정구역'] = df['행정구역'].str.split('(').str.get(0)

df['시'] = df['행정구역'].str.split(' ').str.get(0)

df['구'] = df['행정구역'].str.split(' ').str.get(1)

df['동'] = df['행정구역'].str.split(' ').str.get(2)

df_busan = df[df['시'] == '부산광역시']
```

```python
df_busan
```

```python
df_busan.dtypes
```

```python
def remove_comma(x):
    return x.replace(',', '')
```

```python
df_busan['총인구수'] = df_busan['총인구수'].apply(remove_comma)
```

```python
df_busan['남자'] = df_busan['남자'].apply(remove_comma)
```

```python
df_busan['여자'] = df_busan['여자'].apply(remove_comma)
```

```python
df_busan
```

```python
######5번
df_busan = df_busan.astype({'총인구수':'int'})
```

```python
df_busan = df_busan.astype({'남자':'int'})
```

```python
df_busan = df_busan.astype({'여자':'int'})
```

```python
df_busan.dtypes
```

```python
df_busan['동'].replace('', np.nan, inplace=True)
df_busan.dropna(subset=['동'], inplace=True)
```

```python
df_busan['동']
```

```python
def remove_Jea(x):
    return x.replace('제', '')
```

```python
df_busan['동'] = df_busan['동'].apply(remove_Jea)
```

```python
df_busan['동'].sample(6)
```

```python
df2 = df_busan.groupby(['동']).sum()
```

```python
######6번
map = folium.Map([35.20218639653045, 129.08021906767706],zoom_start=10)
rfile = open('HangJeongDong_ver20220309.geojson','r',encoding ='utf-8').read()
bdata= json.loads(rfile)
jsondata_busan = {'type': 'FeatureCollection'}
jsondata_pick = []
for item in bdata['features']:
    if item['properties']['sidonm'] == '부산광역시':
        item['id'] =  item['properties']['emd_nm'].replace('제','')
        jsondata_pick.append(item)
jsondata_busan['features']=jsondata_pick
c = folium.Choropleth(jsondata_busan,data=df2['총인구수'],columns=[df2.index,df2['총인구수']],key_on='feature.id',fill_color='YlGn',fill_opacity=0.7,line_opacity=0.3,legend_name='부산 인구').add_to(map)
c.geojson.add_child(
    folium.features.GeoJsonTooltip(['emd_nm'],labels=False)
)
map
```
